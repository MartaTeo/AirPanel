```python
import pandas as pd
import matplotlib.pyplot as plt

def hist1(df, col, bins=10, title=None):
    s = df[col].dropna()
    plt.figure(figsize=(6,4))
    if pd.api.types.is_numeric_dtype(s):
        plt.hist(s, bins=bins, edgecolor="white")
        plt.ylabel("Count")
    else:
        s.value_counts().plot(kind="bar")
        plt.ylabel("Count")
        plt.xticks(rotation=0)
    plt.title(title or col)
    plt.tight_layout()
    plt.show()

def hist_by(df, x, by, bins=10, title=None):
    d = df[[x, by]].dropna()
    plt.figure(figsize=(6,4))
    for g, sub in d.groupby(by):
        plt.hist(sub[x], bins=bins, alpha=0.65, label=str(g), edgecolor="white")
    plt.xlabel(x)
    plt.ylabel("Count")
    plt.title(title or f"{x} by {by}")
    plt.legend(title=by)
    plt.tight_layout()
    plt.show()
```

```python
hist1(interviews, "gender")          
hist1(interviews, "age", bins=12)       
hist_by(interviews, "age", "gender", bins=12)  
```

```python
import numpy as np
transport = interviews['location.mobility_access'].unique()
car = []
public = []
own = []
bike = []
for t in transport:
    if 'car' in t:
        car.append(t)

    if 'public' in t:
        public.append(t)

    if 'own' in t:
        own.append(t)

    if 'bike' in t or 'bicycle' in t:
        bike.append(t)

print(f'cars: {car}')
print('_'*35)
print(f'public transportation: {np.array(public)}')
print('_'*35)
print(f'owners: {own}')
print('_'*35)
print(f'bikes: {np.array(bike)}')
```
